```python
# converting the memory columns to SSD, HDD, FLash Memory, Hybrid... columns and dropping the "Memory" column.
```

```python

# Create separate columns for memory types (adding individual memory types first)
raw_df['ssd'] = raw_df['Memory'].str.extract(r'(\d+)GB\s+SSD', expand=False).fillna(0).astype(int)
raw_df['hdd'] = raw_df['Memory'].str.extract(r'(\d+)GB\s+HDD', expand=False).fillna(0).astype(int)
raw_df['flash_storage'] = raw_df['Memory'].str.extract(r'(\d+)GB\s+Flash\s+Storage', expand=False).fillna(0).astype(int)
raw_df['hybrid'] = raw_df['Memory'].str.extract(r'(\d+\.?\d*)TB\s+(?:Hybrid|HDD)', expand=False).fillna(0).astype(float) * 1024  # Convert TB to GB

# Handle sum of different memory types
raw_df['ssd'] += raw_df['Memory'].str.extract(r'(\d+)GB\s+SSD\s*\+\s*(\d+)GB', expand=False).fillna(0).astype(int).sum(axis=1)
raw_df['hdd'] += raw_df['Memory'].str.extract(r'(\d+)GB\s+HDD\s*\+\s*(\d+)GB', expand=False).fillna(0).astype(int).sum(axis=1)
raw_df['flash_storage'] += raw_df['Memory'].str.extract(r'(\d+)GB\s+Flash\s+Storage\s*\+\s*(\d+)GB', expand=False).fillna(0).astype(int).sum(axis=1)
raw_df['hybrid'] += raw_df['Memory'].str.extract(r'(\d+\.?\d*)TB\s+(?:Hybrid|HDD)\s*\+\s*(\d+)GB', expand=False).fillna(0).astype(float).sum(axis=1) * 1024  # Convert TB to GB

# one of the data is hybrid data and is in GB only, we need to capture it too.
raw_df['hybrid'] += raw_df['Memory'].str.extract(r'(\d+)GB\s+Hybrid',expand = False).fillna(0).astype(int)
# Drop the original 'Memory' column
raw_df.drop('Memory',axis = 1, inplace = True)
```

The pattern (r'(\d+)GB\s+SSD') captures the numeric value before "GB" followed by "SSD" in the 'Memory' column.

The '.fillna(0)' function replaces any missing values (NaN) with 0 or when the cell doesnot contain such matching data.
The pattern (r'(\d+\.?\d*)TB\s+(?:Hybrid|HDD)\s*\+\s*(\d+)GB') captures two groups:
the first group captures the numeric value before "TB" (allowing for a decimal point), and the second group captures the numeric value before "GB".

The (?:Hybrid|HDD) part ensures that either "Hybrid" or "HDD" is matched after the memory size.
+= sign adds the changes to the existing column.

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
# fitting the encoder on the those categorical features.
raw_df['ScreenResolution'] = le.fit_transform(raw_df['ScreenResolution'])
raw_df['Cpu'] = le.fit_transform(raw_df['Cpu'])
raw_df['Gpu'] = le.fit_transform(raw_df['Gpu'])
```

```python
x = raw_df.loc[:, raw_df.columns != 'Price'] #all except the price column will be in the x dataset.
y = raw_df['Price']
```

# Feature Selection


```python
from sklearn.feature_selection import SelectFromModel
from sklearn.tree import DecisionTreeRegressor
c=x.copy()
d =y.copy()

# Create a decision tree model
model = DecisionTreeRegressor()

# Perform feature selection based on feature importance
feature_selector = SelectFromModel(model)
X_selected = feature_selector.fit_transform(c, d)

# Get the selected feature indices
selected_feature_indices = feature_selector.get_support(indices=True)

# Get the selected feature names
selected_features = c.columns[selected_feature_indices]

# Print the selected features
print(selected_features)
```

 we can see that using Decision Tree Regressior model of feature selection we got the following features:
 
'Inches','ScreenResolution', 'Cpu', 'Ram', 'Gpu', 'Weight', 'ssd', 'TypeName_Notebook'

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import GradientBoostingRegressor

e =x.copy()
f =y.copy()
# Create a Gradient Boosting model
gb_model = GradientBoostingRegressor()

# Perform feature selection with threshold
threshold = 0.01  # Adjust the threshold value as per your requirement
sfm = SelectFromModel(gb_model, threshold=threshold)
X_selected = sfm.fit_transform(e, f)

# Get the selected feature names
selected_feature_indices = sfm.get_support(indices=True)
selected_feature_names = e.columns[selected_feature_indices]
print("Selected Features:")
print(selected_feature_names)
```

```python
# lasso regression for features selection
## For feature Selection :
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
g = x.copy()
h=y.copy()
# Then we use selectFromModel object from sklearn, which will select features whose coefficients are non-zero
feature_selected_model = SelectFromModel(Lasso(alpha = 0.005, random_state=0)) #use same randam_state for test data
feature_selected_model.fit(g,h)
```
